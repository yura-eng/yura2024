# While-Listening Activity: Learning New Words
2. Gradio TTS App

```python
!pip install gradio gtts

from gtts import gTTS
import gradio as gr

# TTS function using gTTS
def text_to_speech(text):
    """Convert text to speech using gTTS and save as an audio file."""
    output_file = "output.mp3"
    tts = gTTS(text)
    tts.save(output_file)
    return output_file

# Sample text
text = """
The Mystery of the Disappearing Diamond

Characters: Detective Lee, Janitor Amy, Night Guard Mike, Mrs. Black (art collector)
Setting: Inside the museum, after the diamond was stolen

Detective Lee: I can not believe it. The Emerald Star diamond was stolen last night. What happened here?

Mike (Night Guard): I was on duty, but the security alarm was deactivated. I do not know how it happened. I did not hear anything.

Detective Lee: The glass case was broken, and fingerprints were found, but they do not match anyone here. Sam, you were cleaning this area last night, right?

Amy (Janitor): Yes, I was. But I saw nothing suspicious. I was in the next room when the cameras were turned off.

Detective Lee: Strange. Mrs. Black, you were seen at the museum earlier. Why did you leave in such a hurry?

Mrs. Black: I had a meeting. I left before the diamond was stolen. Besides, I would not risk it. I only collect art legally.

Detective Lee: A piece of fabric was found on the display case. It looks like it came from a fancy suit, like the one you are wearing, Mrs. Black.

Mrs. Black: That is impossible! It must belong to someone else.

Amy (Janitor): Actually, my jacket was torn last night too. It could be mine.

Detective Lee: And what about the empty champagne glass? It was left near the display case. Mike, do you know where it came from?

Mike (Night Guard): No, but I saw Mrs. Black holding a glass earlier in the evening.

Mrs. Black: That is a lie! I did not drink anything here.

Detective Lee: Hmm, this is a real puzzle. The alarm was disabled, the cameras were turned off, and the glass case was broken. Someone here is hiding something.
"""

# Gradio Interface
interface = gr.Interface(
    fn=text_to_speech,
    inputs=gr.Textbox(lines=20, value=text, label="Enter Text for TTS"),
    outputs=gr.Audio(label="Generated Audio"),
    title="Mystery Story TTS App",
    description="Enter the text you want to convert to speech. The app will generate an audio file that you can play or download."
)

# Launch the Gradio app
interface.launch()
```

3. Gradio Cloze Question App

```python
!pip install gradio

import gradio as gr

# Cloze question function
def cloze_question(input_text, user_answers):
    """
    Checks user's answers for a Cloze test against the correct answers.
    input_text: The text with blanks represented as '[...]'.
    user_answers: User-provided answers to fill in the blanks.
    """
    # Example input and correct answers
    correct_answers = ["diamond", "alarm", "Mrs. Black", "fabric"]
    feedback = []

    # Split user answers and check correctness
    user_answers = user_answers.split(",")  # Assume answers are comma-separated
    for i, user_answer in enumerate(user_answers):
        if i < len(correct_answers):
            if user_answer.strip().lower() == correct_answers[i].lower():
                feedback.append(f"Blank {i+1}: Correct!")
            else:
                feedback.append(f"Blank {i+1}: Incorrect. Correct answer: {correct_answers[i]}")
        else:
            feedback.append(f"Blank {i+1}: No more blanks to fill!")

    # Final feedback as joined string
    return "\n".join(feedback)

# Example text with blanks
example_text = """
The Mystery of the Disappearing Diamond

Last night, the precious Emerald Star [_________] was stolen. The security [_________] was mysteriously disabled, and no one heard anything. Detective Lee suspects [_________], especially after finding a piece of torn [_________] near the scene.
"""

# Gradio interface
interface = gr.Interface(
    fn=cloze_question,
    inputs=[
        gr.Textbox(lines=5, value=example_text, label="Cloze Text (Read only)", interactive=False),
        gr.Textbox(lines=1, placeholder="Enter your answers, separated by commas", label="Your Answers")
    ],
    outputs=gr.Textbox(label="Feedback"),
    title="Cloze Question App",
    description="Fill in the blanks in the text by entering your answers separated by commas. Click Submit to see if you're correct!",
)

# Launch the app
interface.launch()
```

# After-Listening Activity: Sequence the Story
4. Gradio Image Sequencing App


```python
import gradio as gr
from gtts import gTTS

# Story sections with dialogue and descriptions
story_sections = [
    ("Detective Lee: I can not believe it. The Emerald Star diamond was stolen last night. What happened here?",
     "Detective Lee is speaking, in a museum, after the theft."),
    ("Mike (Night Guard): I was on duty, but the security alarm was deactivated. I do not know how it happened. I did not hear anything.",
     "Mike explains his situation."),
    ("Detective Lee: The glass case was broken, and fingerprints were found, but they do not match anyone here. Sam, you were cleaning this area last night, right?",
     "Detective Lee questions Sam."),
    ("Amy (Janitor): Yes, I was. But I saw nothing suspicious. I was in the next room when the cameras were turned off.",
     "Amy responds to Detective Lee."),
    ("Detective Lee: Strange. Mrs. Black, you were seen at the museum earlier. Why did you leave in such a hurry?",
     "Detective Lee questions Mrs. Black."),
    ("Mrs. Black: I had a meeting. I left before the diamond was stolen. Besides, I would not risk it. I only collect art legally.",
     "Mrs. Black denies involvement."),
    ("Detective Lee: A piece of fabric was found on the display case. It looks like it came from a fancy suit, like the one you are wearing, Mrs. Black.",
     "Detective Lee confronts Mrs. Black."),
    ("Mrs. Black: That is impossible! It must belong to someone else.",
     "Mrs. Black protests."),
    ("Amy (Janitor): Actually, my jacket was torn last night too. It could be mine.",
     "Amy suggests it might be her jacket."),
    ("Detective Lee: And what about the empty champagne glass? It was left near the display case. Mike, do you know where it came from?",
     "Detective Lee questions Mike about the glass."),
    ("Mike (Night Guard): No, but I saw Mrs. Black holding a glass earlier in the evening.",
     "Mike mentions Mrs. Black with the glass."),
    ("Mrs. Black: That is a lie! I did not drink anything here.",
     "Mrs. Black denies drinking."),
    ("Detective Lee: Hmm, this is a real puzzle. The alarm was disabled, the cameras were turned off, and the glass case was broken. Someone here is hiding something.",
     "Detective Lee suspects someone is hiding the truth."),
]

# Create TTS audio files for each section
audio_files = []
for i, (desc, _) in enumerate(story_sections):
    tts = gTTS(desc)
    audio_file = f"audio_{i+1}.mp3"
    tts.save(audio_file)
    audio_files.append(audio_file)

# Function to check if the sequence is correct
def check_order(a, b, c, d):
    user_order = [a.upper(), b.upper(), c.upper(), d.upper()]
    correct_order = ['A', 'B', 'C', 'D']
    return "Correct!" if user_order == correct_order else "Try again."

# Create Gradio interface using Blocks
with gr.Blocks() as iface:
    gr.Markdown("**The Mystery of the Disappearing Diamond**")
    gr.Markdown("Sequence the story parts in the correct order based on the dialogues.")

    # Display the audio and dialogue texts for each part
    for label, (desc, _) in zip(['A', 'B', 'C', 'D'], story_sections):
        gr.Markdown(f"**{label}**: {desc}")
        gr.Audio(value=audio_files[story_sections.index((desc, _))])  # Display TTS audio

    with gr.Row():
        a = gr.Textbox(label="Enter label for the first part of the story")
        b = gr.Textbox(label="Enter label for the second part of the story")
        c = gr.Textbox(label="Enter label for the third part of the story")
        d = gr.Textbox(label="Enter label for the fourth part of the story")

    btn = gr.Button("Check Order")
    output = gr.Textbox(label="Result")

    btn.click(fn=check_order, inputs=[a, b, c, d], outputs=output)

iface.launch()
```

# Writing Activity
5.Gradio Writing Checker App


```python

!pip install language_tool_python

import gradio as gr
import language_tool_python

# LanguageTool 객체 생성 (영어)
tool = language_tool_python.LanguageTool('en-US')

# 문법 및 스타일 체크 함수
def check_grammar(text):
    matches = tool.check(text)
    if not matches:
        return "No mistakes found!"
    else:
        corrected_text = language_tool_python.utils.correct(text, matches)
        return f"Found {len(matches)} mistakes. Here is the corrected text:\n\n{corrected_text}"

# Gradio 인터페이스
with gr.Blocks() as iface:
    gr.Markdown("# Gradio Writing Checker App")
    gr.Markdown("This app checks the grammar and style of the text you input.")

    # 사용자 입력 텍스트 박스
    input_text = gr.Textbox(label="Enter your text here", placeholder="Type your text...", lines=6)

    # 결과 출력
    output = gr.Textbox(label="Corrected Text", placeholder="Your corrected text will appear here.", lines=6)

    # 문법 체크 버튼
    btn = gr.Button("Check Grammar")

    # 버튼 클릭 시 문법 체크 실행
    btn.click(fn=check_grammar, inputs=input_text, outputs=output)

iface.launch()
```
